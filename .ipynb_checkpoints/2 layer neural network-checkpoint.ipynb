This is a 2-layer neural network implemented using the sigmoid activation function. It is done completely from scratch in Python. Its performance is similar to that of an MLP Regressor in Scikit-learn.

I implemented this neural network as a part of my learning expereince in CSE 840 - Computational Foundations of Artificial Intelligence, using Linear Algebra learned in the class.


- load the S&P-500 dataset into a dataframe object.
- Compute the change in S&P-500 index between every two consecutive trading days.
- Create a vector $\textbf{y}$ that contains the change in S&P-500 index for each day and a corresponding matrix $\mathbf{X}$ that contains changes in S&P-500 index in each of its 5 previous days. 
- Display the matrix $\mathbf{X}$ and vector $\mathbf{y}$ obtained along with their sizes.


```python
import pandas as pd
import numpy as np

sp500 = pd.read_csv('sp500.csv', header='infer')
N = sp500.shape[0]
temp = sp500['SP-500'].values
sp500_diff = pd.DataFrame(sp500['Date'][:N-1])
sp500_diff['Diff'] = temp[:N-1] - temp[1:N]
sp500_diff


N = sp500_diff.shape[0]
data = sp500_diff.loc[:N-6,['Date','Diff']]
data['diff(t-1)'] = sp500_diff.loc[1:N-5,'Diff'].values
data['diff(t-2)'] = sp500_diff.loc[2:N-4,'Diff'].values
data['diff(t-3)'] = sp500_diff.loc[3:N-3,'Diff'].values
data['diff(t-4)'] = sp500_diff.loc[4:N-2,'Diff'].values
data['diff(t-5)'] = sp500_diff.loc[5:N-1,'Diff'].values
data


y = data['Diff'].values
X = data.drop(columns=['Date','Diff']).values 
print('y =', y)
print('X =', X)

X.shape
```

 Divide the data into training and test sets, using 67% of the data for training and the remaining 33% for testing. 

```python
from sklearn.model_selection import train_test_split

X_train, X_test, Y_train, Y_test =  train_test_split(X, y, test_size=0.33, random_state=1)
print('Training set size for predictors:', X_train.shape)
print('Test set size for predictors:', X_test.shape)
```

**Step 2:** Model Fitting with a 2-Layer Neural network

<img src="network.png" width=500px>

Below is the 2-layer neural network with an architecture that is shown in the figure above. Each neuron in the hidden layer will generate an output $a_k$ according to the following formula:
$$a_k = \sigma_1\bigg( \sum_j w_{kj} x_j + w_{k0}\bigg)$$
where $\sigma_1(z) = \frac{1}{1 + e^{-z}}$ is the sigmoid activation function, $x_j$ is the $j$-th attribute value of the input data, $w_{kj}$ is the weight of the connection between $x_j$ and the $k$-th hidden neuron of the network, whereas $w_{k0}$ is the bias term of the $k$-th hidden neuron. In contrast, the neuron in the output layer will generate the predicted output of the neural network using the equation below:
$$\hat{y} = \sum_k v_{k} a_k + v_{0}$$
where $v_k$ is the weight of the link between the $k$-th hidden neuron and the output neuron and $v_0$ is the bias term of the output neuron. 

**Note that the output layer uses a linear activation function, i.e., $\sigma_2(z) = z$, since the network is used to solve a regression problem.**\

The weights of the model is updated using the stochastic gradient descent algorithm. Specifically, the weights are immediately updated for each input example $(\mathbf{x} \in \mathbb{R}^d,y \in \mathbb{R})$ in each iteration as follows:
\begin{eqnarray}
w_{kj} \leftarrow w_{kj} - \alpha \frac{\partial \ell}{\partial w_{kj}} \ \ \ \textrm{and} \ \ \ 
v_{k} \leftarrow v_{k} - \alpha \frac{\partial \ell}{\partial v_{k}}
\end{eqnarray}
where the loss function is given by $\ell = (y - \hat{y})^2$. Using chain rule, the partial derivative of the loss function with respect to each model parameter can be compute as follows:
$$\frac{\partial \ell}{\partial w_{kj}} = \frac{\partial \ell}{\partial a_{k}} \times \frac{\partial a_{k}}{\partial z_{k}} \times \frac{\partial z_{k}}{\partial w_{kj}}, \ \ \ \textrm{and} \ \ \ \frac{\partial \ell}{\partial v_{k}} = \frac{\partial \ell}{\partial \hat{y}} \times \frac{\partial \hat{y}}{\partial v_{k}}$$
where $z_k = \sum_j w_{kj} x_j + w_{k0}$. The resulting partial derivatives are summarized below:
\begin{eqnarray}
\frac{\partial \ell}{\partial w_{kj}} &=& 2(y - \hat{y})v_k a_k(1-a_k) x_j, \\ 
\frac{\partial \ell}{\partial w_{k0}} &=& 2(y - \hat{y})v_k a_k(1-a_k) \\
\frac{\partial \ell}{\partial v_{k}} &=& 2 (y - \hat{y}) a_k, \\
\frac{\partial \ell}{\partial v_{0}} &=& 2 (y - \hat{y}) 
\end{eqnarray}

```python
from tqdm import tqdm
import matplotlib.pyplot as plt

class NN():

    def __init__(self, input_dim, hidden_dim = 2, random_state=1):
        """        
            Input:  
                input_dim:    dimension of its input attributes
                hidden_dim:   dimension of the hidden layer (i.e., number of hidden neurons)
                random_state: seed for random number generator.
 
            This function should create the following variables:
            - self.W: a matrix of size hidden_dim x input_dim to represent the weight parameter of the hidden layer
            - self.b: a vector of length hidden_dim x 1 to represent the bias parameter of the neurons in the hidden layer
            - self.V: a vector of size hidden_dim x 1 to represent the weight parameter of the links to the output layer
            - self.V0: a scalar to represent the bias parameter of the neuron in the output layer.
            
            You should initialize self.W and self.V randomly from a Gaussian distribution with mean 0 and 
            standard deviation equals to 0.1. The bias terms, self.b and self.V0 are initialized to zeros.
            You should also set the seed of the random number generator to the random_state variable.
            
        """    
        np.random.seed(random_state)
        self.W = np.random.normal(loc=0.0, scale=0.1, size=(hidden_dim, input_dim))
        self.b = np.zeros((hidden_dim, 1))
        self.V = np.random.normal(loc=0.0, scale=0.1, size=(hidden_dim, 1))
        self.V0 = 0.0
         
    def forward(self, x):
        """
            This function computes the forward propagation step of a layer in the neural network. 
            
            Input:  
                x: a vector whose length is equal to input_dim

            Output:
                a:    a vector whose length is equal to hidden_dim.
                yhat: predicted value given by the output layer of the network. 

        """        
        z = np.dot(self.W, x.T) + self.b
        a = 1 / (1 + np.exp(-z))
        yhat = np.dot(self.V.T, a) + self.V0
        return a, yhat
    
    def backprop(self, x, y, a, yhat):
        """
            This function performs the backward propagation step of the neural network. Specifically,
            it calculates the gradient of the loss function with respect to the weight and bias parameters
            and returns them as grad_W, grad_b, grad_V, and grad_V0, respectively.
            
            Input:  
                x: input vector.
                y: output scalar (ground truth).
                a: output activation of the hidden layer
                yhat: current prediction generated by output layer of the network.

            Output:
                grad_W, grad_b, grad_V, grad_V0: gradients of the loss with respect to the model parameters 
                
        """
        # Compute error in output layer
        delta = y - yhat
        grad_V = -(2*delta * a)
        grad_V0 = -(2 * delta)
        temp = (a * (1 - a))
        grad_W = -2*delta*self.V*temp*x
        grad_b = -2*(delta * self.V * temp)
        
        
        print("grad_W", grad_W.shape)
        print("W", self.W.shape)

        return grad_W, grad_b, grad_V, grad_V0

    
    def fit(self, X, Y, max_epochs = 20, learning_rate = 0.001, plot_loss = False): 
        """
            This function is invoked when training the neural network model.
        
            Input:
                X: A 2-d numpy array consisting of the predictor attribute values of the training data  
                Y: A numpy array vector consisting of the target values of the training data
                max_epochs: maximum number of epochs for backpropagation algorithm
                learning_rate: learning rate for backpropagation algorithm
                plot_loss: A True/False value. It will display the root-mean-square error at each epoch 
                           if set to True. 
            Output:
                None
        """       
        losses = []
        for epoch in tqdm(range(max_epochs), 'epoch'):
            epoch_loss = 0
            for x, y in zip(X, Y):
                a, yhat = self.forward(x.reshape(1, -1))
                grad_W, grad_b, grad_V, grad_V0 = self.backprop(x.reshape(1, -1), y, a, yhat) 
                self.W -= learning_rate * grad_W
                self.b -= learning_rate * grad_b
                self.V -= learning_rate * grad_V
                self.V0 -= learning_rate * grad_V0
                epoch_loss += (yhat - y) ** 2
                
            epoch_loss = np.sqrt(epoch_loss / len(X))
            losses.append(np.sqrt(((yhat - y) ** 2).mean()))

            
        if plot_loss:
            plt.plot(losses)
            plt.xlabel('epoch')
            plt.ylabel('Loss (rmse)')
            plt.title('loss')
            plt.show()
        
    def predict(self, X):
        """
        The function is used to predict the output values for a set of data points X. Specifically,
            for each data point, it will call the forward() function to calculate the predicted value
            based on the output of the last layer of the neural network. 

            Input:
                X: A 2-dimensional numpy array of predictor attribute values. 

            Output:
                Y_pred: A numpy array that contains the predicted values of the input data.
        """
        a, y_pred = self.forward(X)
        return y_pred
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Create a 2-layer neural network with 20 hidden neurons
nn = NN(input_dim=X_train.shape[1], hidden_dim=20, random_state=1)

learning_rate = 0.00001
max_epochs = 200

# Fit the model to the input training data
nn.fit(X_train, Y_train, max_epochs=max_epochs, learning_rate=learning_rate, plot_loss=False)
```

```python
# Compute the mean squared error of the predictions
y_pred = nn.predict(X_test)
print(y_pred)
rmse = np.sqrt(((Y_test - y_pred) ** 2).mean())
print(f" Root Mean squared error: {rmse:.4f}")
```
